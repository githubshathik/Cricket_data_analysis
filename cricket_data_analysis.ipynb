```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
f=df.sort_values(by ='strikerate',ascending = False).head(10)
f
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5), dpi=150)
plt.bar(f['batsman'],f['strikerate'])
plt.xlabel("batsman")
plt.ylabel("strikerate")
plt.title('Top 10 Strike rates')
plt.legend(["STRIKERATES"])
plt.xticks(rotation=90)
plt.show()
```

```python
a=df.sort_values(by ='total_runs',ascending = False).head(10)
a
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5), dpi=150)
plt.bar(a['batsman'],a['total_runs'])
plt.xlabel("batsman")
plt.ylabel("total_runs")
plt.title('Top 10 most runs')
plt.legend(["runs"])
plt.xticks(rotation=90)
plt.show()
```
